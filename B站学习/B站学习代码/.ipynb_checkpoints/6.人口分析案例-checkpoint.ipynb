```python
import numpy as np
import pandas as pd
from pandas import DataFrame
```

```python
#导入文件，查看文件
abb = pd.read_csv('state-abbrevs.csv') #state(州的全称)abbreviation(州的简称)
abb
```

```python
area = pd.read_csv('state-areas.csv') #state（州的全称）area（sq，mi）州的面积
area
```

```python
pop = pd.read_csv('state-population.csv') #state/region简称，ages年龄，year时间，population人口
pop
```

```python
#将人口数据和各州简称数据进行合并
abb_pop = pd.merge(abb,pop,left_on='abbreviation',right_on='state/region',how='outer')
abb_pop.head()
```

```python
#将合并的数据中重复的abbreviation列删除
abb_pop.drop(labels='abbreviation',axis=1,inplace=True)
```

```python
abb_pop.head()
```

```python
#查看缺失数据的列
#方式1：isnull,notll,any,all
abb_pop.isnull().any(axis=0)
#state,population这两列中存在空值
```

```python
#方式2：
abb_pop.info()
```

```python
#找到的这些state/region的state的值为NaN，进行去重操作（将state中的空值对应的简称找到，且对简称进行去重）
abb_pop.head()
```

```python
#思路:可以将state这一列中的空值对应的行数据取出，从该行数据中就可以取出简称的值
```

```python
#1.将state中的空值定位到
abb_pop['state'].isnull()
#2.将上述的布尔值作为源数据的行索引
abb_pop.loc[abb_pop['state'].isnull()] #将state中空对应的行数据取出
#3.将简称去重
abb_pop.loc[abb_pop['state'].isnull()]['state/region'].unique()

#结论：只有PR和USA对应的全称数据为空值
```

```python
abb_pop.loc[abb_pop['state'].isnull()]
```

```python
#为找到的这些state/region的state项补上正确的值，从而去除掉state这一列的所有NaN
#思考：填充该需求中的空值可不可以使用fillna?
    # - 不可以。fillna可以使用空的紧邻值做填充。fillna（value=‘xxx’）使用指定的值填充空值
    # 使用给元素赋值的方式填充！
```

```python
#1.先给USA的全称对应的空值进行批量赋值
```

```python
#1.1将USA对应的行数据找出（行数据中就存在state的空值）
abb_pop['state/region'] == 'USA'
abb_pop.loc[abb_pop['state/region'] == 'USA'] #将USA对应的行数据取出
```

```python
#1.2将USA对应的全称空对应的行索引取出
indexs = abb_pop.loc[abb_pop['state/region'] == 'USA'].index
```

```python
abb_pop.iloc[indexs]
abb_pop.loc[indexs,'state'] = 'United States'
```

```python
#2.可以将PR的全称进行赋值
abb_pop['state/region'] == 'PR' #PR对应的行数据
abb_pop.loc[abb_pop['state/region'] == 'PR']
```

```python
indexs = abb_pop.loc[abb_pop['state/region'] == 'PR'].index
abb_pop.iloc[indexs]
abb_pop.loc[indexs,'state'] = 'Puerto Rico'
```

```python
#合并各州面积数据areas
abb_pop_area = pd.merge(abb_pop,area,on='state',how='outer')
```

```python
#我们会发现area(sq.mi)这一列有缺失数据，找出是哪些行
abb_pop_area['area (sq. mi)'].isnull()
indexs = abb_pop_area.loc[abb_pop_area['area (sq. mi)'].isnull()].index  #空值对应的行数据
```

```python
#去除含有缺失数据的行
abb_pop_area.drop(labels=indexs,axis=0,inplace=True)
```

```python
#找出2010年的全民人口数据(基于df做条件查询)
abb_pop_area['year'] == '2010'
```
