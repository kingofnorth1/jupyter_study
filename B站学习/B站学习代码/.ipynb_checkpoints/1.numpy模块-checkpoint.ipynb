将外部一张图片加载到numpy数组中，然后改变数组元素

```python
import matplotlib.pyplot as plt
img_arr =  plt.imread('./黄芯桐.jpg') #返回的数组，数组中装载的就是图片内容
plt.imshow(img_arr) #将numpy数组进行可视化
```

```python
img_arr = img_arr - 100 #将每一个数组元素都减去一个100
plt.imshow(img_arr)
```
