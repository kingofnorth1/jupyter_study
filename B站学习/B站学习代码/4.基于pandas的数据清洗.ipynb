处理丢失数据
- 有两种丢失数据：
    - None
    - np.nan(NaN)
    
    
- 两种丢失数据的区别

```python
import numpy as np
type(None) #None对象类型
```

```python
type(np.nan) #NAN浮点型
```

- 为什么在数据分析中需要用到的是浮点类型的空而不是对象类型
    - 数据分析中常常使用某些形式的运算来处理原始数据，如果原数据中的空值为NAN形式，则不会干扰或者中断运算
    - NAN可以参加运算的
    - None是不可以参与运算

- 在pandas中如果遇到了None形式的空值则pandas会将其强转成NAN的形式

```python
np.nan + 1
```

- pandas处理空值操作
    - isnull
    - notnull
    - any
    - all
    - dropna
    - fillna

```python
import pandas as pd
from pandas import DataFrame,Series
```

```python
#伪造一组数据(存在空值)
df = DataFrame(data=np.random.randint(0,100,size=(8,6)))
df.iloc[2,3] = None
df.iloc[4,4] = np.nan
df.iloc[5,2] = None
df
```

- 方式1：对空值进行过滤（删除空所在的行数据）
    - 技术：isnull,notnull,any,all

```python
df.isnull()
```

```python
#那些行中存在true
#any:用来检测行或者列中是否存在True
df.isnull().any(axis=1)
#将上部的布尔值作为源数据的行索引
df.loc[df.isnull().any(axis=1)] #true对应的行数据就是存在缺失值的行数据
drop_index = df.loc[df.isnull().any(axis=1)].index #即将要删除的行索引
df.drop(labels=drop_index,axis=0)
```

```python
df.notnull().all(axis=1)  #any遇到true返回true，all遇到false返回false
df.loc[df.notnull().all(axis=1)]
```

 - 规律：
    - isnull->any
    - notnull->all

- 方式2：
    - dropna:可以直接将缺少的行或者列进行删除

```python
df.dropna(axis=0)
```

- 对缺失值进行覆盖
    - fillna

```python
df.fillna(method='ffill',axis=1)  #使用水平方向的向前填充去填充空值
```

```python
df.loc[:,[1,2,3,4]]
```

```python
df.dropna(axis=0) #这里的1代表列
```

```python
df.notnull().all(axis=1) #这里的1代表行
df.loc[df.notnull().all(axis=1)]
```

```python
#填充空值
data = df.fillna(method='ffill',axis=0).fillna(method='bfill',axis=0) #0代表列
```

```python
#检查下是否填充完整
data.isnull().any(axis=0) #0代表列
```

- 处理异常数据
    - 自定义一个1000行3列（A,B,C）取值范围为0-1的数据源，然后将C列中的值大于其两倍标准差的异常值进行清洗

```python
df = DataFrame(data=np.random.random(size=(1000,3)),columns=['A','B','C'])
#判定异常值的条件
twice_std = df['C'].std() * 2

df['C'] < twice_std
df.loc[df['C'] < twice_std]
```
