```python
import pandas as pd
df = pd.read_csv('in_data/weather.csv')
temp = df[ (df['city'] == 'B') & (df['temperature'] < 999990)]['temperature'].max()
df[(df['city'] == 'B') & (df['temperature'] == temp)]
```

```python
def func(ser):
    rain08 = ser['rain08']
    rain20 = ser['rain20']
    if rain08 == 999990 or rain20 == 999990:
        return 999990
    else:
        return rain20 + rain08

df.apply(func, axis = 1)
```

```python
df.drop( df[df["temperature"] >= 999990].index )
```

```python
for key, gdata in df.groupby(['city', 'date']):
    missing = gdata [ gdata['rain08'] == 999990]
    ok = gdata [ gdata['rain08'] != 999990]
    if len(ok) == 0:
        df.loc [ missing.index, 'rain08'  ] = 0
    elif len(missing) != 0:
        mean = ok['rain08'].mean()
        df.loc [ missing.index, 'rain08' ] = mean
```
