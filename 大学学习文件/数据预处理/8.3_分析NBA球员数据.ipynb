```python
import pandas as pd
df = pd.read_csv('in_data/harden18-19.csv')
def process(x):
    isdigits = list(map(str.isdigit, x))
    st = isdigits.index(True, 2)
    return x[:st]
df['比赛'] = df['比赛'].apply(process)
df
```

```python
df['助攻标准化'] = (df['助攻']-df['助攻'].min())/(df['助攻'].max()-df['助攻'].min())
df['得分标准化'] = (df['得分']-df['得分'].min())/(df['得分'].max()-df['得分'].min())
df['篮板标准化'] = (df['篮板']-df['篮板'].min())/(df['篮板'].max()-df['篮板'].min())
df['发挥'] = df['助攻标准化'] + df['得分标准化'] + df['篮板标准化']
df.sort_values(by=['发挥'],ascending=False).head()
```

```python
groups1=pd.cut(df['得分'], 5)
groups2=pd.qcut(df['得分'], 5)
#使用下面语句看不同区间的统计值
pd.value_counts(groups1).sort_index()
pd.value_counts(groups2).sort_index()
```

```python
import numpy as np
pd.pivot_table(df, values=['助攻','得分','篮板'], index=['比赛', '果'],
aggfunc=[np.average], fill_value=0)
```
