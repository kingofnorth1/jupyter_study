## 2.1 使用PCA降维算法进行降维数据处理

### 1. 原始数据2维可视化

```python
import matplotlib.pyplot as plt
import matplotlib

# 指定中文字体
myfont = matplotlib.font_manager.FontProperties(
	fname='D:/jupyter file/date/msyh.ttc'
)

# 代码中未使用。但3.2版本以下的matplotlib绘制3d图形需要引入
import mpl_toolkits.mplot3d 

from sklearn import datasets
from sklearn.decomposition import PCA

# 导入数据
iris = datasets.load_iris()
X = iris.data[:, :2]  # 只拿前两列特征，绘制数据分布图
y = iris.target

x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5

plt.figure(2, figsize=(8, 6))
plt.clf()

# 绘制图形，
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1, edgecolor="k")
plt.xlabel("花瓣长度",fontproperties=myfont)
plt.ylabel("花瓣宽度",fontproperties=myfont)

# 设置图形的轴范围
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
```

### 2. 原始数据4维数据降成3维后可视化

```python
fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d", elev=-150, azim=110)

# 使用全部4列数据，将4维特征降为3维
# n_components表示保留几维数据
X_reduced = PCA(n_components=3).fit_transform(iris.data)
ax.scatter(
    X_reduced[:, 0],
    X_reduced[:, 1],
    X_reduced[:, 2],
    c=y,
    cmap=plt.cm.Set1,
    edgecolor="k",
    s=40,
)

ax.set_title("PCA 3维数据特征",fontproperties=myfont)
ax.set_xlabel("1st PCA向量",fontproperties=myfont)
ax.xaxis.set_ticklabels([])
ax.set_ylabel("2nd PCA向量",fontproperties=myfont)
ax.yaxis.set_ticklabels([])
ax.set_zlabel("3rd PCA向量",fontproperties=myfont)
ax.zaxis.set_ticklabels([])

plt.show()
```

### 任务： 原始数据使用PCA降成2维换后进行可视化：

```python
fig = plt.figure(1, figsize=(8, 6))

# 使用全部4列数据，将4维特征降为2维
# n_components表示保留几维数据
X_reduced = PCA(n_components=2).fit_transform(iris.data)
plt.scatter(
    X_reduced[:, 0],
    X_reduced[:, 1],
    c=y,
    cmap=plt.cm.Set1,
    edgecolor="k",
    s=60,
)

plt.xlabel("特征一",fontproperties=myfont)
plt.ylabel("特征二",fontproperties=myfont)


plt.show()
```
