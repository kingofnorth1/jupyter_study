```python
from sklearn.datasets import load_boston
import joblib
import pandas as pd
```

```python
boston_dataset = load_boston()
boston = pd.DataFrame(boston_dataset.data)
boston.columns = boston_dataset.feature_names
#定义Y标签，即房子的价格
boston["PRICE"] = boston_dataset.target
```

```python
boston.head()
```

```python
boston.describe()
```

```python
import seaborn as sns
sns.pairplot(boston,vars=['DIS','RAD','TAX','PTRATIO','B','LSTAT','PRICE'])
```

```python
sns.pairplot(boston,vars=['DIS','RAD','TAX','PTRATIO','B','LSTAT','PRICE'])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(boston.drop('PRICE',axis=1),boston['PRICE'],\
                                              test_size=0.3,random_state=0)
```

```python
X_train.head()
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(X_train,y_train)
```

```python
y_pred = LR.predict(X_test)
```

```python
from sklearn import metrics
mse = metrics.mean_squared_error(y_test,y_pred)
print("MSE=",round(mse,4))
print("wO=",round(LR.intercept_,2))
print("W=",[round(X,2)for X in LR.coef_])
```
