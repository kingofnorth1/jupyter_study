## 3 散点图

## 3.1 引入所需库

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 3.2 数据准备

```python
# 数据准备
N = 1000
x = np.random.randn(N) #随机生成1000个[0,1)之间的数据
y = np.random.randn(N) #随机生成1000个[0,1)之间的数据|
```

## 3.3 用 Matplotlib 画散点图

```python
# 用 Matplotlib 画散点图
plt.scatter(x, y,marker='>') #点的标识为右箭头
plt.show()
```

## 3.4 用  Seaborn 画散点图

```python
df = pd.DataFrame({'x': x, 'y': y}) #转换数据
sns.jointplot(x="x", y="y", data=df, kind='scatter') #生成散点图
plt.show()
```

## 4 折线图

## 4.1 引入所需库

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 4.2 数据准备

```python
# 数据准备
x = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
y = [5, 3, 6, 20, 17, 16, 19, 30, 32, 35]
```

# 5 直方图

## 5.1 引入所需要的库

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 5.2 数据准备

```python
# 数据准备
a = np.random.randn(100) #随机生成1000个[0,1)之间的数据
s = pd.Series(a) #转换数据
```

## 5.3 绘制直方图

```python
# 用 Matplotlib 画直方图
plt.hist(s) #绘制直方图
plt.show() #展示图表
# 用 Seaborn 画直方图
sns.displot(s, kde=False) #不使用密度函数
plt.show() #展示图表
sns.displot(s, kde=True) #使用密度函数
plt.show() #展示图表
```

# 7 箱线图

# 10 雷达图

## 10.1 引入所需库

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

from pylab import mpl
# 设置显示中文字体
mpl.rcParams["font.sans-serif"] = ["SimHei"]
# 设置正常显示符号
mpl.rcParams["axes.unicode_minus"] = False
```

## 10.2 数据准备

```python
zhfont = mpl.font_manager.FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-microhei.ttc')
# 数据准备
labels=np.array([" 推进 ","KDA"," 生存 "," 团战 "," 发育 "," 输出 "])
stats=[83, 61, 95, 67, 76, 88]
# 画图数据准备，角度、状态值
angles=np.linspace(0, 2*np.pi, len(labels), endpoint=False)#数据准备
stats=np.concatenate((stats,[stats[0]]))#数值聚合
angles=np.concatenate((angles,[angles[0]]))#角度聚合
labels=np.concatenate((labels,[labels[0]]))#标签聚合
```

## 10.3 绘制雷达图

```python
# 用 Matplotlib 画蜘蛛图
fig = plt.figure()#新建画布
ax = fig.add_subplot(111, polar=True)#创建极坐标
ax.plot(angles, stats, 'o-', linewidth=2)#绘制空心圆直线
ax.fill(angles, stats, alpha=0.25)#填充数据
ax.set_thetagrids(angles * 180/np.pi, labels)#绘制网格线
plt.show()#显示图表
```

# 11 二元变量分布图

## 11.1 引入所需库

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## 11.2 数据准备

```python
#下载数据http://i9000.net:8888/sgn/HUP/Visual/3/tips.csv至运行目录
```

```python
# 数据准备
tips = pd.read_csv("./tips.csv")#读取数据
print(tips.head(10))#展示前10条数据
```

```python
# 用 Seaborn 画二元变量分布图（散点图，核密度图，Hexbin 图）
sns.jointplot(x="total_bill", y="tip", data=tips, kind='scatter')
sns.jointplot(x="total_bill", y="tip", data=tips, kind='kde')
sns.jointplot(x="total_bill", y="tip", data=tips, kind='hex')
plt.show()
```
