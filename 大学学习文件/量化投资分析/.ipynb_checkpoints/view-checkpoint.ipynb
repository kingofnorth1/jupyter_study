<h1>导入必要的包

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = ['SimHei']
plt.rcParams['font.size'] = 12
from pymysql import Connect
```

<h1>连接数据库

```python
CONN_PARAM = dict(host="localhost", user="root", password="123456",
                  database="sakila_dwh", charset="utf8")
conn = Connect(**CONN_PARAM)
```

<h1>设置配色以及方法定义

```python
colors = ['#614099', '#8481BA', '#B4B4D5', '#1663A9', '#4995C6', '#92C2DD']
def view1(sql, x, y, s):
    cur = conn.cursor()
    cur.execute(sql)
    data = cur.fetchall()
    # 截取数据，只取排名前10的国家
    data = list(sorted(data, key=lambda x: x[1], reverse=True))[:9]
    x_data = []
    y_data = []

    for i in data:
        if i[0] == "" or i[0] is None:
            continue
        x_data.append(str(i[x]))
        y_data.append(i[y])
    plt.figure(figsize=(8, 6))
    bars = plt.barh(x_data, y_data, color=colors)
    for bar in bars:
        height = bar.get_height()
    # plt.text(bar.get_x() + bar.get_width()/2., height, int(height), ha='center', va='bottom')
    plt.title(s)
    plt.xticks(rotation=45)
    plt.savefig(s+".png")
    plt.show()
```

```python
def view3(sql, x, y, s):
    cur = conn.cursor()
    cur.execute(sql)
    data = cur.fetchall()
    x_data = []
    y_data = []
    for i in data:
        if i[0] == "" or i[0] is None:
            continue
        x_data.append(i[x])
        y_data.append(i[y])
    plt.figure(figsize=(8, 6))
    plt.pie(y_data, labels=x_data, colors = colors, autopct='%1.1f%%')
    plt.title(s)
    plt.savefig(s+".png")
    plt.show()
```

<h1>创收排名前五的电影营业额

```python
sql = """SELECT a.film_title, ROUND(sum(b.rental_amount), 2) num FROM dim_film a, fact_rental b
    WHERE a.film_key = b.film_key GROUP BY a.film_title"""
cur = conn.cursor()
cur.execute(sql)
data = cur.fetchall()
# 截取数据，只取排名前5的电影
data = list(sorted(data, key=lambda x: x[1], reverse=True))[:5]
x_data = []
y_data = []
for i in data:
    x_data.append(i[0])
    y_data.append(i[1])
plt.figure(figsize=(8, 6))
bars = plt.bar(x_data, y_data, color=colors)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height, int(height), ha='center', va='bottom')
plt.title('创收排名前五的电影营业额')
plt.xticks(rotation=45)
plt.show()
```

<h1>统计最长租赁时间和最短租赁时间的碟片

```python
sql = """SELECT
 	sto.store_name,
--  	f.title,
--  	f.rental_duration
	count(f.rental_duration)
FROM
	film f
left join inventory inv on f.film_id = inv.film_id
left join store sto on sto.store_id = inv.store_id
where f.rental_duration = (select min(rental_duration) from film)
GROUP BY sto.store_name"""
view2(sql, 0, 1, '各个租赁店的最长租赁时间')
```

```python
sql = """SELECT
 	sto.store_name,
--  	f.title,
--  	f.rental_duration
	count(f.rental_duration)
FROM
	film f
left join inventory inv on f.film_id = inv.film_id
left join store sto on sto.store_id = inv.store_id
where f.rental_duration = (select max(rental_duration) from film)
GROUP BY sto.store_name"""
view2(sql, 0, 1, '各个租赁店的最短租赁时间')
```

<h1>每个店的租赁收入，每个客户的收入，每店/每客户

```python
sql = """SELECT
 	sto.store_name,
--  	f.title,
--  	f.rental_duration
	count(f.rental_duration)
FROM
	film f
left join inventory inv on f.film_id = inv.film_id
left join store sto on sto.store_id = inv.store_id
where f.rental_duration = (select max(rental_duration) from film)
GROUP BY sto.store_name"""
view2(sql, 0, 1, '每个店的收入')
```

```python
sql = """SELECT
 	sto.store_name,
	pay.customer_id,
	concat(cust.first_name,' ',cust.last_name) as name,
	sum(amount) as amout
--   (select sum(amount) from payment pay left join customer cust on pay.customer_id = cust.customer_id GROUP BY cust.store_id)/sum(amount)
FROM
	payment pay
left join customer cust on pay.customer_id = cust.customer_id
left join store sto on cust.store_id = sto.store_id
WHERE sto.store_name="Store-A"
GROUP BY pay.customer_id
ORDER BY amout DESC
LIMIT 0,10

"""
view1(sql, 2, 3, "Store-A前十个客户的总收入")
```

```python
sql = """SELECT
 	sto.store_name,
	pay.customer_id,
	concat(cust.first_name,' ',cust.last_name) as name,
	sum(amount) as amout
--   (select sum(amount) from payment pay left join customer cust on pay.customer_id = cust.customer_id GROUP BY cust.store_id)/sum(amount)
FROM
	payment pay
left join customer cust on pay.customer_id = cust.customer_id
left join store sto on cust.store_id = sto.store_id
WHERE sto.store_name="Store-B"
GROUP BY pay.customer_id
ORDER BY amout DESC
LIMIT 0,10

"""
view1(sql, 2, 3, "Store-B前十个客户的总收入")
```

<h1>每天平均收入，每店，每客户，每店/每客户组合

```python
sql = """SELECT
	date_format(payment_date, '%Y/%m/%d') AS Date,
	sum(amount),
	sto.store_id
FROM
	payment pay
LEFT JOIN staff sta ON sta.staff_id = pay.staff_id
LEFT JOIN store sto ON sta.store_id = sto.store_id
WHERE sto.store_id=1
GROUP BY
	Date,
	sto.store_id"""
view1(sql, 0, 1 , 'Store-A每天平均收入')
```

```python
sql = """SELECT
	date_format(payment_date, '%Y/%m/%d') AS Date,
	sum(amount),
	sto.store_id
FROM
	payment pay
LEFT JOIN staff sta ON sta.staff_id = pay.staff_id
LEFT JOIN store sto ON sta.store_id = sto.store_id
WHERE sto.store_id = 2
GROUP BY
	Date,
	sto.store_id"""
view1(sql, 0, 1 , 'Store-B每天平均收入')
```

<h1>每个职员出租的影片数量，客户数量，和时长

```python
sql = """SELECT
	sta.first_name,
	count(inv.film_id),
	count(distinct ren.customer_id),
	sum(UNIX_TIMESTAMP(return_date)-UNIX_TIMESTAMP(rental_date))/(60*60*24)
FROM
	staff sta
left join rental ren on sta.staff_id = ren.staff_id
left join inventory inv on ren.inventory_id = inv.inventory_id
group by sta.staff_id
"""
view2(sql, 0, 1, '每个职员出租的影片数量')
```
