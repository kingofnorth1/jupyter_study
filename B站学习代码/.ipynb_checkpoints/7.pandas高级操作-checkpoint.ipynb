# 替换操作
    - 单值替换：
        - 替换复合要求的元素：to_replace=15,value='e'
        - 将指定列的元素进行替换to_replace={列索引：被替换的值}
    - 多值替换：
        - 列表替换：to_replace=[],value=[]
        - 字典替换：to_replace={to_replace:valule,to_replace:value}

```python
df = DataFrame(data=np.random.randint(0,100,size=(5,6)))
df
```

```python
df.replace(to_replace=2,value='Two')
```

```python
df.replace(to_replace={1:'one'})
```

```python
df.replace(to_replace={4:5},value='five')
#将指定列的元素进行替换to_replace={列索引：被替换的值}
```

映射操作
- 概念：创建一个映射关系列表，把values元素和一个特定的标签或者字符串绑定（给一个元素值提供不同的表现形式）

```python
dic = {
    'name':['张三','李四','张三'],
    'salary':[15000,20000,15000]
}
df = DataFrame(data=dic)
df
```

```python
#映射关系表
dic = {
    '张三':'tom',
    '李四':'jack'
}
df['e_name'] = df['name'].map(dic)
df
```

- map是Series的方法，只能被Series调用

# 运算工具
- 超过3000部分的钱缴纳50%的税，计算每个人的税后薪资

```python
#该函数是我们指定的一个运算法则
def after_sal(s): #计算s对映的税后薪资
    return s - (s-3000)*0.5
    
df['after_sal'] = df['salary'].map(after_sal) #可以将df['salary']这个Series中每一个元素作为参数
df
```

排序实现的随机抽样

# 数据分类处理
  - 数据分类处理的核心：
      - groupby()函数
      - groups属性查看分组情况

```python
df = DataFrame({'item':['Apple','Banana','Orange','Banana','Orange','Apple'],
               'price':[4,3,3,2.5,4,2],
               'color':['red','yellow','yellow','green','green','green'],
               'weight':[12,20,50,30,20,44]})
df
```

```python
#想要水果的种类进行分析
df.groupby(by='item')
```

```python
#查看详细的分组情况
df.groupby(by='item').groups
```

- 分组聚合

```python
#计算出每一种水果分组情况
df.groupby(by='item')['price'].mean()
```

```python
#计算每一种颜色对应水果的平均重量
df.groupby(by='color')['weight'].mean()
```

```python
dic = df.groupby(by='color')['weight'].mean().to_dict()
#将计算出的平均重量汇总到源数据
df['mean_w'] = df['color'].map(dic)
df
```
