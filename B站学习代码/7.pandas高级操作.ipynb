# 替换操作
    - 单值替换：
        - 替换复合要求的元素：to_replace=15,value='e'
        - 将指定列的元素进行替换to_replace={列索引：被替换的值}
    - 多值替换：
        - 列表替换：to_replace=[],value=[]
        - 字典替换：to_replace={to_replace:valule,to_replace:value}

```python
df = DataFrame(data=np.random.randint(0,100,size=(5,6)))
df
```

```python
df.replace(to_replace=5,value='Two')
```

```python
df.replace(to_replace={14:'one'})
```

```python
df.replace(to_replace={4:10},value='five')
#将指定列的元素进行替换to_replace={列索引：被替换的值}
```

## 映射操作
- 概念：创建一个映射关系列表，把values元素和一个特定的标签或者字符串绑定（给一个元素值提供不同的表现形式）

```python
dic = {
    'name':['张三','李四','张三'],
    'salary':[15000,20000,15000]
}
df = DataFrame(data=dic)
df
```

```python
#映射关系表
dic = {
    '张三':'tom',
    '李四':'jack'
}
df['e_name'] = df['name'].map(dic)
df
```

- map是Series的方法，只能被Series调用

# 运算工具
- 超过3000部分的钱缴纳50%的税，计算每个人的税后薪资

```python
#该函数是我们指定的一个运算法则
def after_sal(s): #计算s对映的税后薪资
    return s - (s-3000)*0.5
    
df['after_sal'] = df['salary'].map(after_sal) #可以将df['salary']这个Series中每一个元素作为参数
df
```

# 数据分类处理
  - 数据分类处理的核心：
      - groupby()函数
      - groups属性查看分组情况

```python
df = DataFrame({'item':['Apple','Banana','Orange','Banana','Orange','Apple'],
               'price':[4,3,3,2.5,4,2],
               'color':['red','yellow','yellow','green','green','green'],
               'weight':[12,20,50,30,20,44]})
df
```

```python
#想要水果的种类进行分析
df.groupby(by='item')
```

```python
#查看详细的分组情况
df.groupby(by='item').groups
```

- 分组聚合

```python
#计算出每一种水果分组情况
df.groupby(by='item')['price'].mean()
```

```python
#计算每一种颜色对应水果的平均重量
df.groupby(by='color')['weight'].mean()
```

```python
dic = df.groupby(by='color')['weight'].mean().to_dict()
#将计算出的平均重量汇总到源数据
df['mean_w'] = df['color'].map(dic)
df
```

# 高级数据聚合
   - 使用groupby分组后，也可以使用transform和apply提供自定义函数更多的运算
   - df.groupby('item)['price'].sum() <==> df.groupby('item)['price'].apply(sum)
   - transform和apply都会进行运算，在transform或者apply中传入函数即可

```python
def my_mean(s):
    m_sum = 0
    for i in s:
        m_sum += i
    return m_sum / len(s)
```

```python
df.groupby(by='item')['price'].transform(my_mean)
```

```python
df.groupby(by='item')['price'].apply(my_mean)
```

# 透视表
  - 透视表是一种可以对数据动态排布并且分类汇总的表个格式。在pandas中被称作pivot_table

```python
import pandas as pd
import numpy as np
df = pd.read_csv('./B站学习数据/透视表-篮球赛.csv',encoding='utf-8')
df 
```

# pivot_table有四个最重要的参数index、values、columns、Aggfunc

- 查看哈登对阵同一个对手在不同主客场下的数据，分类条件为对手和主客场

```python
df.pivot_table(index=['对手','主客场'])
```

- values参数：需要对计算的数据进行筛选
    - 如果我们只是需要哈登在主客场和不同胜负情况下的得分、篮板与助攻三项数据：

```python
df.pivot_table(index=['主客场','胜负'],values=['得分','篮板','助攻'])
```

- Aggfunc参数：设置我们对数据聚合时进行的函数操作
    - 当我们未设置aggfunc时，它默认是aggfunc='mean'计算均值。
    
- 还想获得james harden在主客场和不同胜负情况下的总得分、总篮板、总助攻时:

```python
df.pivot_table(index=['主客场','胜负'],values=['得分','篮板','助攻'],aggfunc='sum')
```

- Columns:可以设置列层次字段
    - 对values字段进行分类

```python
#获取所有队主客场的总得分
df.pivot_table(index='主客场',values=['得分'],aggfunc='sum')
```

```python
#获取每个队主客场的总得分(在总得分的基础上又进行了对手的分类)
df.pivot_table(index='主客场',values='得分',columns='对手',aggfunc='sum',fill_value=0)
```

## 交叉表
  - 是一种用于计算分组的特殊透视表，对数据进行汇总
  - pd.crosstab(index,columns)
      - index:分组数据，交叉表的行索引
      - columns：交叉表的列索引

```python
import pandas as pd
from pandas import DataFrame
```

```python
df = DataFrame({'sex':['man','man','women','women','man','women','man','women','women'],
               'age':[15,23,25,17,35,57,24,31,22],
               'smoke':[True,False,False,True,True,False,False,True,False],
               'height':[168,179,181,166,173,178,188,190,160]})
df
```

- 求出各个性别抽烟的人数

```python
pd.crosstab(df.smoke,df.sex)
```

- 求出各个年龄段抽烟人情况

```python
pd.crosstab(df.age,df.smoke)
```
