```python
import pandas as pd
```

```python
df = pd.read_excel('data/人民币存贷款利率.xlsx')
```

```python
df.head()
```

```python
df.index = df.iloc[:,0].values
```

```python
df = df.resample('Y').mean()
df['年'] = df.index
df['年'] = df['年'].apply(lambda x : str(x).split('-')[0])
df.index = df['年']
df = df.drop('年',axis=1)
```

```python
df.head()
```

```python
df.to_excel('数据/人民币存贷款利率.xlsx')
```
