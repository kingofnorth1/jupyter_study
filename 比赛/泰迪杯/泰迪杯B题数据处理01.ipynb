```python
import seaborn as sns   #读取数据
import pandas as pd
%matplotlib  inline
y = pd.read_csv('./附件1-区域15分钟负荷数据.csv')
indu = pd.read_csv('./附件2-行业日负荷数据.csv')
tianqi = pd.read_csv('./附件3-气象数据.csv')
tianqi = tianqi.drop([0],axis=0)
tianqi
```

```python
#del tianqi['Unnamed: 6']  #删除Unnamed：6
```

```python
tianqi['最高温度'] = tianqi['最高温度'].map(lambda d: d.replace('℃','')).astype(int)  #使用空格替换摄氏度
tianqi['最低温度'] = tianqi['最低温度'].map(lambda d: d.replace('℃','')).astype(int) 
```

```python
series = tianqi.join(tianqi['天气状况'].str.split('/',expand=True))  #以/切割天气状况并且分配给天气1与天气2
tianqi['天气1'] = series[0]
tianqi['天气2'] = series[1]
tianqi.info()
```

```python
tianqi['白天风力风向'].unique()  #去重
```

```python
tianqi['夜晚风力风向'].unique()
```

```python
dic = { '无持续风向微风':0, 
        '无持续风向1-2级':1,        
        '东南风1-2级':1,
        '南风1-2级':1,
        '东风1-2级':1,
        '北风1-2级':1,
        '东北风1-2级':1,
        '西南风1-2级':1,
        '东北偏东风2':2, 
        '微风<3级':3, 
        '无持续风向<3级':3,
        '北风3':4,
        '东北风3～4级':5, 
        '东南风3-4级':5,
        '东南风3～4级':5,
        '东风3-4级':5,
        '北风3～4级':5,
        '南风3～4级':5, 
        '南风3-4级':5, 
        '东风3～4级':5, 
        '西南风3-4级':5,
        '东北风3-4级':5, 
        '北风3-4级':5,
        '北风4-5级':6, 
        '北风4～5级':6,
        '南风4～5级':6,
        '东南风4-5级':6, 
        '东风8-9级':7,}  #使用字典映射
tianqi['白天风力风向'] = tianqi['白天风力风向'].map(dic)
tianqi['夜晚风力风向'] = tianqi['夜晚风力风向'].map(dic)
```

```python
dic1 = { '晴':0, 
         '晴间多云':1,
         '局部多云':2,
         '多云':3,
         '雾':4,
         '阴':5, 
         '阵雨':6,
         '雷阵雨':7,
         '小雨':8,
         '小到中雨':9,
         '小雨-中雨':10, 
         '中雨':11,
         '中雨-大雨':12,
         '大雨':13,
         '大到暴雨':14,
         '暴雨':15
       }
tianqi['天气1'] = tianqi['天气1'].map(dic1)  #使用去重后数据映射
tianqi['天气2'] = tianqi['天气2'].map(dic1)
del tianqi['天气状况']
```

```python
y = y.rename(columns={'数据时间':'日期1'})  #更改列名称
y['日期'] = y['日期1'].apply(lambda x: x.split(' ')[0])  #使用split分隔日期，把年月日赋值给日期
tianqi.loc[:, '日期'] = pd.to_datetime(tianqi.loc[:, '日期'], format='%Y年%m月%d日', errors='coerce')  #将日期规格化
y.loc[:, '日期'] = pd.to_datetime(y.loc[:, '日期'], format='%Y/%m/%d', errors='coerce') #将日期规格化
train = y.merge(tianqi,on='日期',how='left')  #将y表与tianqi表合并
del train['日期']  #删除日期列
```

```python
train.to_csv("泰迪杯数据01.csv",index=False,encoding='utf_8_sig')  #将数据写入
```
