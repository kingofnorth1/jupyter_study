```python
import pandas as pd
import numpy as np
import openpyxl
```

```python
sheet1 = pd.read_excel(r"D:\jupyter file\数据处理-日常\1.xlsx", sheet_name='Sheet1');
sheet2 = pd.read_excel(r"D:\jupyter file\数据处理-日常\1.xlsx", sheet_name='Sheet2');
```

```python
sheet1.columns = ['calc_month', 'start_months', 'periods', 'amount']
```

```python
sheet1
```

```python
# 转换日期列
sheet1['calc_month'] = pd.to_datetime(sheet1['calc_month'], format='%Y-%m')
sheet1['start_month'] = pd.to_datetime(sheet1['start_months'], format='%Y-%m')
```

```python
# 创建一个函数来找到6个月的总金额
def calculate_num_month_total(df, start_month, periods, monthss):
    # 获取6个月内的数据
    end_month = start_month + pd.DateOffset(months=monthss)
    filtered_data = df[
        (df['start_month'] == start_month) &
        (df['calc_month'] < end_month) &
        (df['periods'] == periods)
    ]
    
    # 计算总金额
    total_amount = filtered_data['amount'].sum()
    return total_amount
```

```python
sheet2['交易月'] = pd.to_datetime(sheet2['交易月'], format='%Y-%m')
```

```python
# 应用到Sheet2的数据上
sheet2['6个月金额合计'] = sheet2.apply(
    lambda row: calculate_num_month_total(sheet1, row['交易月'], row['期数'], 8), axis=1
)
```

```python
# 应用到Sheet2的数据上
sheet2['12个月金额合计'] = sheet2.apply(
    lambda row: calculate_num_month_total(sheet1, row['交易月'], row['期数'], 14), axis=1
)
```

```python
# 应用到Sheet2的数据上
sheet2['18个月金额合计'] = sheet2.apply(
    lambda row: calculate_num_month_total(sheet1, row['交易月'], row['期数'], 20), axis=1
)
```
